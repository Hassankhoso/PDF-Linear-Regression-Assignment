```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Advertising Budget and Sales.xls')
df.head()
```

```python
df.info()
```

```python
plt.title('Sales By TV Advertising')
sns.distplot(df['Sales ($)'])
plt.show()
```

```python
plt.scatter(df['TV Ad Budget ($)'], df['Sales ($)'], color = 'lightcoral')
plt.title('Sales Vs Tv Ad Budget')
plt.xlabel('Tv Ad Budget')
plt.ylabel('Sales')
plt.box(False)
plt.show()
```

```python
X = df[['TV Ad Budget ($)']]  
y = df['Sales ($)'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train) 
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Tv Ad Budget Vs Sales (Training Set)')
plt.xlabel('Tv Ad Budget')
plt.ylabel('Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Tv Ad Budget/Sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Tv Ad Budget Vs Sales (Test Set)')
plt.xlabel('Tv Ad Budget')
plt.ylabel('Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Tv Ad Budget/Sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
